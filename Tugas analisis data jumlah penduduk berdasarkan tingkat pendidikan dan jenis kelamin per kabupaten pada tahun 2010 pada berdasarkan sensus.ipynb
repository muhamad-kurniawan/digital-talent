## Deskripi Data

Data jumlah penduduk berdasarkan tingkat pendidikan dan jenis kelamin per kabupaten pada tahun 2010 pada berdasarkan sensus

## Statistik tentang Data


```python
import pandas as pd

data_pendidikan=pd.read_csv('processedjumlahpenduduktingkatpendidikanjeniskelaminkabupatensensus2010.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_pendidikan
```

```python
# jumlah penduduk berdasarkan tingkat pendidikan dan jenis kelamin per provinsi
data_pendidikan.groupby(['nama_provinsi','tingkat_pendidikan']).sum()
```

```python
# jumlah penduduk berdasarkan tingkat pendidikan dan jenis kelamin se-Indonesia
data_pendidikan.groupby('tingkat_pendidikan').sum()
```

## Visualisasi Sederhana


```python
ax = data_pendidikan.groupby('tingkat_pendidikan').mean()['jumlah_laki_laki']
ax = ax.plot(kind='bar', title ="Jummlah Laki laki terhadap tingkat pendidikan", legend=True, fontsize=12)
ax.set_xlabel("tingkat_pendidikan", fontsize=12)
ax.set_ylabel("jumlah_laki_laki", fontsize=12)
```

```python
ax = data_pendidikan.groupby('tingkat_pendidikan').mean()['jumlah_perempuan']
ax = ax.plot(kind='bar', title ="Jummlah perempuan terhadap tingkat pendidikan", legend=True, fontsize=12)
ax.set_xlabel("tingkat_pendidikan", fontsize=12)
ax.set_ylabel("jumlah_perempuan", fontsize=12)
```
